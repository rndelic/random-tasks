```python
def gcd(a, b):
    if b > a:
        return gcd(b, a)
    else:
        while b != 0:
            temp = b
            b = a % b
            a = temp
        return a

    
def foldl(f, x0, lst):
    if not lst:
        return x0
    return foldl(f, f(x0, lst[0]), lst[1:])


def gcd_n(numbers):
    return foldl(gcd, 0, numbers)
        
    
def count_max_candies(candy_array):
    count_dict = {}
    for candy in candy_array:
        if candy not in count_dict.keys():
            count_dict[candy] = 1
        else:
            count_dict[candy] += 1
    return gcd_n(list(count_dict.values()))
```

```python
count_max_candies(['a', 'a', 'a', 'b', 'b', 'b', 'b', 'c', 'c', 'c', 'c', 'c'])
```
